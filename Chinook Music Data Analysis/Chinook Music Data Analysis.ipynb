```python
## counries with the most invoices


ps.sqldf("""
        select billing_country, count(*) as number_of_invoices
        from invoice
        group by billing_country
        order by number_of_invoices desc
""")
```

```python
## Top five cities for invoice


ps.sqldf("""
        select billing_city, sum(total) as invoice_total
        from invoice
        group by billing_city
        order by invoice_total desc
        limit 5
""")
```

```python
## top  5 customers with their spendings

ps.sqldf("""
        select c.customer_id, c.first_name, c.last_name, sum(i.total) as total_spending
        from customer c
        join invoice i
         on c.customer_id = i.customer_id
            group by c.customer_id
            order by total_spending desc
        limit 5
""")
```

```python
## All classical music listeners details


ps.sqldf("""

    select distinct email as Email_id, first_name as FirstName, last_name as Last_Name, g.name as genre
    from customer c
    join invoice i on i.customer_id = c.customer_id
    join invoice_line il on il.invoice_id = i.invoice_id
    join track t on t.track_id = il.track_id
    join genre g on g.genre_id = t.genre_id
    where g.name like "Classical"
    order by email
""")
```

```python
## All Sci-Fi Fantasy deatils

ps.sqldf("""
        select distinct email as Email_id, first_name as FirstName, last_name as LastName, g.name as genre
        from customer c
        join invoice i on i.customer_id = c.customer_id
        join invoice_line il on il.invoice_id = i.invoice_id
        join track t on t.track_id = il.track_id
        join genre g on g.genre_id = t.genre_id
        where g.name like "Rock"
        order by email

""")
```

```python
## Sales for each country with respect to genre

ps.sqldf("""
    select count(*) As sales_per_genre, g.name, g.genre_id, c.country
    from customer c
    join invoice i on i.customer_id = c.customer_id
    join invoice_line il on il.invoice_id = i.invoice_id
    join track t on t.track_id = il.track_id
    join genre g on g.genre_id = t.genre_id
    group by g.name, g.genre_id, c.country
    order by g.name desc
""")
```

```python
## Track length with more then Avg track length and a minute column

ps.sqldf("""
    select name,milliseconds, (milliseconds / 1000) / 60 as Minutes
    from track
    where milliseconds >(
            select avg(milliseconds) as avg_track_length
            from track)
    order by milliseconds desc
    """)
```

```python
## Countries with most tracks purchased

ps.sqldf("""

    select i.billing_country, count(*) tracks_purchased
    from invoice i
        join invoice_line il on i.invoice_id = il.invoice_id
        join track t on t.track_id = il.track_id
        join genre g on g.genre_id = t.genre_id
     group by billing_country
     order by tracks_purchased desc
     limit 10

    """)
```

```python
# Countries with most rock tracks purchased
ps.sqldf("""
select i.billing_country, count(*) Rock_tracks_purchased
from invoice i
    join invoice_line il on i.invoice_id = il.invoice_id
    join track t on t.track_id = il.track_id
    join genre g on g.genre_id = t.genre_id
where g.name = 'Rock'
group by billing_country
order by Rock_tracks_purchased desc
limit 10
""")
```

```python
## countries with most classical tracks purchased

ps.sqldf("""
    select i.billing_country, count(*) Classical_tracks_purchased
    from invoice i
        join invoice_line il on i.invoice_id = il.invoice_id
        join track t on t.track_id = il.track_id
        join genre g on g.genre_id = t.genre_id
     where g.name ='Classical'
     group by billing_country
     order by Classical_tracks_purchased
     limit 10

""")
```

```python
## most albums by a composer

ps.sqldf("""
    select composer, count(album_id) as number_of_album
    from track
    group by composer
    order by number_of_album

    """)
```

```python
## most albums by an artist

ps.sqldf("""
    select count(album_id) as total_album, at.name
    from album a
            join artist at on a.artist_id = at.artist_id
    group by at.name
    order by total_album desc
""")
```
